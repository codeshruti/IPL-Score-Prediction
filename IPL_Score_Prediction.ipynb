```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
```

```python
ipl = pd.read_csv("ipl.csv")
ipl.head()
```

```python
remove_columns = ['striker','non-striker','mid','batsman','bowler','venue','runs_last_5','wickets_last_5	']
ipl.drop(labels=remove_columns,axis=1,inplace=True)
```

Feature Engineering

```python
# Rename KXIP
ipl['bat_team'] = ipl['bat_team'].replace({'Kings XI Punjab':'Punjab Kings'})
ipl['bowl_team'] = ipl['bowl_team'].replace({'Kings XI Punjab':'Punjab Kings'})
ipl['bat_team'] = ipl['bat_team'].replace({'Delhi Daredevils':'Delhi Capitals'})
ipl['bowl_team'] = ipl['bowl_team'].replace({'Delhi Daredevils':'Delhi Capitals'})
```

```python
ipl['bat_team'].unique()
```

```python
current_teams = ['Kolkata Knight Riders','Chennai Super Kings','Mumbai Indians','Punjab Kings','Royal Challengers Bangalore','Delhi Capitals','Sunrisers Hyderabad','Rajasthan Royals']
```

```python
ipl = ipl[(ipl['bat_team'].isin(current_teams))&(ipl['bowl_team'].isin(current_teams))]
```

```python
ipl = ipl[ipl['overs']>=5.0]
```

Data Preprocessing

```python
from datetime import datetime
ipl['date'] = ipl['date'].apply(lambda x:datetime.strptime(x,"%Y-%m-%d"))
```

Encoding

```python
encode_ipl = pd.get_dummies(data=ipl, columns=['bat_team','bowl_team'])
```

```python
encode_ipl.head()
```

```python
encode_ipl = encode_ipl[['date','bat_team_Chennai Super Kings', 'bat_team_Delhi Capitals',
       'bat_team_Punjab Kings', 'bat_team_Kolkata Knight Riders',
       'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Capitals',
       'bowl_team_Punjab Kings', 'bowl_team_Kolkata Knight Riders',
       'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
       'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad','overs','runs', 'wickets',
       'total']]
```

```python
encode_ipl.head()
```

```python
X_train = encode_ipl.drop(labels='total',axis=1)[encode_ipl['date'].dt.year<=2015]
X_test = encode_ipl.drop(labels='total',axis=1)[encode_ipl['date'].dt.year>=2016]
```

```python
Y_train = encode_ipl[encode_ipl['date'].dt.year<=2015]['total'].values
Y_test = encode_ipl[encode_ipl['date'].dt.year>=2016]['total'].values
```

```python
X_train.drop(labels='date',axis=True,inplace=True)
X_test.drop(labels='date',axis=True,inplace=True)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
lasso_regressor = GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X_train,Y_train)
prediction = lasso_regressor.predict(X_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(Y_test, prediction))
print('MSE:', metrics.mean_squared_error(Y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, prediction)))
```

Model for deployment

```python
filename = 'Score-model.pkl'
pickle.dump(lasso_regressor,open(filename,'wb'))
```
